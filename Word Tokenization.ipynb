Assignment 1. Word Tokenization using Python

(1) Tokenization & Remove stopword without Stemming/Lemmatization.

(a) Read text from file carroll-alice.txt‚Äù and print it

```
import urllib.request
url = 'https://raw.githubusercontent.com/rutum/tf-idf/master/gutenberg/carroll-alice.txt'
with urllib.request.urlopen(url) as f:
  text = f.read()
  print(text)
```

(b) Normalize words: convert all upper-case letters to lower case (for example, 'Word' and 'word' are considered as the same word).

```
lowtext = text.lower()
print(lowtext)
```

(c) Tokenization: using nltk word_tokenize.

we converted the html above (a byte-like object) into a string using .decode and tokenized the content.

```
import nltk  #nltk stand for natural language toolkit
nltk.download('punkt')
tokens = nltk.word_tokenize(lowtext.decode('utf-8'))
print(tokens)
```

(d) Remove stopwords using stopwords from nltk. You can add more stopword if needed.


Stop Words are words which do not contain important significance to be used in Search Queries. Usually, these words are filtered out from search queries because they return a vast amount of unnecessary information. 

```
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
filtered_token = [word for word in tokens if not word in stopwords and word.isalnum()]
print(filtered_token)
```

The function would print the occurrence of words and show the top 10 words.

```
def term_count(token_doc):
  text_tf = dict((term, token_doc.count(term)) for term in token_doc)
  print(text_tf)
  freq = nltk.FreqDist(text_tf)
  print("Top 10 freqency word", freq.most_common(10))
  return freq.most_common(10)
```

(e) Count the occurrence of words and display the top 10 common words.

```
initial = term_count(filtered_token)
```


2. Tokenization & Remove stopwords with Stemming/Lemmatization.

PorsterStemmer

Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language.

```
from nltk import PorterStemmer
ps = PorterStemmer()
token_psstem = [ps.stem(word) for word in filtered_token]

stem = term_count(token_psstem)
```

Lemmatization

Lemmatization reduces the inflected words properly ensuring that the root word belongs to the language.

```
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
token_lem = [wnl.lemmatize(word) for word in filtered_token]

lemma = term_count(token_lem)
```

Stemming a word or sentence may result in words that are not actual words. Stems are created by removing the suffixes or prefixes used with a word without knowledge of the context and therefore can't discriminate between words which have different meanings depending on part of speech.

Lemmatization returns an actual word of the language, it is used where it is necessary to get valid words.
In lemmatisation, the part of speech of a word should be first determined and the normalisation rules will be different for different part of speech.

The difference between them explains why 'look' appears in the result of stemming but not in the result of lemmatization.

For example, after stemming, the term 'queer-looking' would be counted as 'look'. However, if we do lemmatization, it won't be counted. Therefore, 'look' appears 102 times in the result with stemming  while 'looked' only appears 45 times in the one with lemmatization.

The difference before/after lemmatization is that it identify a canonical representative for a set of related word forms.

Take 'take' for example, the result without lemmatization would see 'takes' and 'taken' as different words.

In conclusion, lemmatization offers better precision than stemming. However, if the tenses do matter, both of them shouldn't be used.
